Work with the [80Cereal dataset](https://www.kaggle.com/crawford/80-cereals)and find answers to the following questions:

1. Calculate calories per ounce for each cereal product.
2. Identify the product with the highest value of calories per ounce.
3. Identify the product with the lowest value of calories per ounce.

Please submit your solutions as a PDF file to Blackboard by **Wednesday, Sep. 29th at 11:59 PM**.

```python
import numpy as np
import pandas as pd

data_file = pd.read_csv("C:\\Users\\Daiyan\\PycharmProjects"
                        "\\CMP464\\Pandas\\cereal.csv")
high_avg = 0.0
high_product = ''
low_avg = 5000.0
low_product = ''
```

```python
for index, row in data_file.iterrows():
    avg_calories = float(row['calories']) / float(row['weight'])
    if avg_calories > high_avg:
        high_avg = avg_calories
        high_product = str(row['name'])
    if avg_calories < low_avg:
        low_avg = avg_calories
        low_product = str(row['name'])
    print('Average calories per ounce for ' + str(row['name']) + ': '
          + str(avg_calories) + '\n')
```

```python
print("Product with the highest average calories is " + high_product +
      ' with an average calorie count of: ' + str(high_avg) + '\n')
print("Product with the lowest average calories is " + low_product +
      ' with an average calorie count of: ' + str(low_avg))
```
